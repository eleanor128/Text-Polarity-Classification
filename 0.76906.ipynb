# EDA

```python
import pandas as pd
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import string 
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from collections import Counter

nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# 自訂停用字
custom_stopwords = {"s", "n't"}  


def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_stopwords(text, stopwords_list):
    
    text = text.lower()  # 轉小寫
    text = remove_punctuation(text)  # 移除標點符號
    
    return ' '.join([word for word in text.split() if word not in stopwords_list])

def plot_wordcloud_from_words(words, title):
    wordcloud = WordCloud(width=400, height=300, background_color='white').generate_from_frequencies(dict(words))
    plt.figure(figsize=(8, 6))  
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.tight_layout()
    plt.show()

stopwords_combined = stop_words.union(custom_stopwords)

def get_most_common_words(text, n=50):
    word_list = text.split()
    word_counts = Counter(word_list)
    return word_counts.most_common(n)

df_train['TEXT_no_stopwords'] = df_train['TEXT'].apply(lambda x: remove_stopwords(x, stopwords_combined))
text_label_0_no_stopwords = ' '.join(df_train[df_train['LABEL'] == 0]['TEXT_no_stopwords'])
most_common_label_0_no_stopwords = get_most_common_words(text_label_0_no_stopwords, 50)

text_label_1_no_stopwords = ' '.join(df_train[df_train['LABEL'] == 1]['TEXT_no_stopwords'])
most_common_label_1_no_stopwords = get_most_common_words(text_label_1_no_stopwords, 50)

plot_wordcloud_from_words(most_common_label_0_no_stopwords, "Most Common 50 Words in LABEL 0 (Remove Stopwords)")
plt.tight_layout()
plot_wordcloud_from_words(most_common_label_1_no_stopwords, "Most Common 50 Words in LABEL 1 (Remove Stopwords)")

vectorizer = TfidfVectorizer(stop_words='english', max_features=500)
X = vectorizer.fit_transform(df_train['TEXT_no_stopwords'])
```

# Clustering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt

vectorizer = TfidfVectorizer(stop_words='english')  
X = vectorizer.fit_transform(df_train['TEXT'])

k_range = range(2, 11)

# Davies-Bouldin 指數（DBI）來選擇 k
dbi_scores = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    df_train['cluster'] = kmeans.fit_predict(X)
    dbi_score = davies_bouldin_score(X.toarray(), df_train['cluster'])
    dbi_scores.append(dbi_score)

# 繪製 DBI 圖
plt.figure(figsize=(8, 6))
plt.plot(k_range, dbi_scores, marker='o')
plt.title('Davies-Bouldin Index for Different k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Davies-Bouldin Index')
plt.show()

# 假設最好的 k 是 2
best_k = 2
kmeans_best = KMeans(n_clusters=best_k, random_state=42)
df_train['cluster'] = kmeans_best.fit_predict(X)

# PCA 降維可視化
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X.toarray())

# 畫圖
plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df_train['cluster'], cmap='viridis')
plt.title(f'Cluster visualization using PCA (k={best_k})')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar()
plt.show()
```

# RoBERTa

```python
import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from torch.utils.data import DataLoader, TensorDataset
from torch.optim import AdamW
from sklearn.model_selection import KFold
import numpy as np
import optuna

texts = df_train["TEXT"]
labels = df_train["LABEL"]

tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
encodings = tokenizer(texts.tolist(), padding=True, truncation=True, max_length=128, return_tensors='pt')
dataset = TensorDataset(encodings['input_ids'], encodings['attention_mask'], torch.tensor(labels.tolist()))

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 交叉驗證
kf = KFold(n_splits=10, shuffle=True, random_state=42)

class EarlyStopping:
    def __init__(self, patience=2, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.best_loss = np.inf
        self.counter = 0

    def __call__(self, val_loss):
        if val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
        else:
            self.counter += 1

        if self.counter >= self.patience:
            print("Early stopping!")
            return True
        return False

def train(model, train_loader, optimizer, device):
    model.train()
    total_loss = 0
    for batch in train_loader:
        batch = tuple(b.to(device) for b in batch)  # 移動到GPU
        input_ids, attention_mask, labels = batch

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        loss.backward()
        optimizer.step()

    return total_loss / len(train_loader)

def evaluate(model, val_loader, device):
    model.eval()
    total_loss = 0
    correct = 0
    with torch.no_grad():
        for batch in val_loader:
            batch = tuple(b.to(device) for b in batch)  # 移動到GPU
            input_ids, attention_mask, labels = batch

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

            preds = torch.argmax(outputs.logits, dim=1)
            correct += (preds == labels).sum().item()

    return total_loss / len(val_loader), correct / len(val_loader.dataset)

# Optuna 
def objective(trial):

    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 5e-5)
    batch_size = trial.suggest_categorical('batch_size', [8, 16, 32])
    epochs = trial.suggest_int('epochs', 3, 4)

    early_stopping = EarlyStopping(patience=2, min_delta=0.01) 

    model = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=2)
    model.to(device)
    optimizer = AdamW(model.parameters(), lr=learning_rate)

    # K-Fold
    val_accuracies = []
    for fold, (train_idx, val_idx) in enumerate(kf.split(dataset)):
        print(f"Training fold {fold + 1}")
        
        # 切分訓練集和驗證集
        train_data = torch.utils.data.Subset(dataset, train_idx)
        val_data = torch.utils.data.Subset(dataset, val_idx)

        train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_data, batch_size=batch_size)

        # 訓練和驗證
        for epoch in range(epochs):
            print(f"Epoch {epoch + 1}:")
            train_loss = train(model, train_loader, optimizer, device)
            val_loss, val_accuracy = evaluate(model, val_loader, device)

            print(f"  Train Loss: {train_loss}")
            print(f"  Validation Loss: {val_loss}")
            print(f"  Validation Accuracy: {val_accuracy}")

            if early_stopping(val_loss):
                break
        
        val_accuracies.append(val_accuracy)

    return np.mean(val_accuracies)  # 返回驗證集的平均準確度

# 進行 Optuna 優化
study = optuna.create_study(direction='maximize') 
study.optimize(objective, n_trials=3)  

print(f"Best trial: {study.best_trial.params}")
```

```python
# 顯示最佳超參數
best_learning_rate = study.best_trial.params['learning_rate']
best_batch_size = study.best_trial.params['batch_size']
epochs = study.best_trial.params['epochs']
print(f"Best hyperparameters: learning_rate={best_learning_rate}, batch_size={best_batch_size}, epochs={epochs}")

# 初始化模型
best_model = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=2)
best_model.to(device)

# 設置優化器
optimizer = AdamW(best_model.parameters(), lr=best_learning_rate)

# 創建 DataLoader（對整個訓練集進行訓練）
train_loader = DataLoader(dataset, batch_size=best_batch_size, shuffle=True)

# 訓練過程
def final_train(model, train_loader, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total_samples = 0
    
    for batch in train_loader:
        batch = tuple(b.to(device) for b in batch)  # 移動到GPU或CPU
        input_ids, attention_mask, labels = batch

        optimizer.zero_grad()

        # 正向傳播
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits

        total_loss += loss.item()

        # 計算準確度
        preds = torch.argmax(logits, dim=1)
        correct += (preds == labels).sum().item()
        total_samples += labels.size(0)

        # 反向傳播
        loss.backward()
        optimizer.step()

    avg_loss = total_loss / len(train_loader)
    avg_accuracy = correct / total_samples
    return avg_loss, avg_accuracy

# 最終訓練整個訓練集
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")
    train_loss, train_accuracy = final_train(best_model, train_loader, optimizer, device)
    print(f"  Training Loss: {train_loss}")
    print(f"  Training Accuracy: {train_accuracy * 100:.2f}%")

# 保存最終訓練的模型和分詞器
best_model.save_pretrained("./final_model")
tokenizer.save_pretrained("./final_model")
print("Final model and tokenizer have been saved.")
```

# Testing Dataset

```python
import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from torch.utils.data import DataLoader
import pandas as pd
import numpy as np

# 載入最佳的模型和分詞器
final_model = RobertaForSequenceClassification.from_pretrained("./final_model", num_labels=2)
tokenizer = RobertaTokenizer.from_pretrained("./final_model")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
final_model.to(device)


test_texts = df_test["TEXT"].tolist()  
encodings = tokenizer(test_texts, padding=True, truncation=True, max_length=128, return_tensors='pt')

test_dataset = torch.utils.data.TensorDataset(encodings['input_ids'], encodings['attention_mask'])
test_loader = DataLoader(test_dataset, batch_size=16)


def predict(final_model, test_loader, device):
    final_model.eval()  # 設置模型為評估模式
    all_preds = []
    with torch.no_grad():
        for batch in test_loader:
            batch = tuple(b.to(device) for b in batch)  
            input_ids, attention_mask = batch

            # 模型預測
            outputs = final_model(input_ids, attention_mask=attention_mask)
            logits = outputs.logits

            preds = torch.argmax(logits, dim=1)
            all_preds.extend(preds.cpu().tolist())

    return all_preds


predictions = predict(final_model, test_loader, device)

# 儲存結果
df_test['LABEL'] = predictions

df_test.head()
```
